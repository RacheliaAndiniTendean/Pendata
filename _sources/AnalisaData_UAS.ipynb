# Analisis Data Menggunakan Algoritma Random Forest dan K-Nearest Neighbor (KNN) pada Dataset Higher Education Students Performance Evaluation

## 1. Pendahuluan

Data mining merupakan salah satu teknik yang digunakan untuk menemukan pola atau informasi penting dari sekumpulan data. Salah satu penerapan data mining adalah klasifikasi, yaitu proses mengelompokkan suatu data ke dalam kelas tertentu berdasarkan data yang telah dipelajari sebelumnya.

Pada analisis ini digunakan dataset **Higher Education Students Performance Evaluation** yang berasal dari UCI Machine Learning Repository. Dataset tersebut berisi data evaluasi mahasiswa yang mencakup beberapa faktor, seperti karakteristik mahasiswa, mata kuliah yang diambil, aktivitas belajar, hingga nilai akhir yang diperoleh.

Analisis dilakukan menggunakan dua algoritma klasifikasi, yaitu **Random Forest** dan **K-Nearest Neighbor (KNN)**. Kedua algoritma tersebut dipilih karena cukup sering digunakan dalam proses klasifikasi dan memiliki cara kerja yang berbeda. Random Forest melakukan prediksi dengan membangun banyak pohon keputusan, sedangkan KNN melakukan klasifikasi berdasarkan kedekatan jarak antar data.

Proses analisis dilakukan menggunakan aplikasi **KNIME Analytics Platform**, mulai dari tahap preprocessing data, pembagian data latih dan data uji, proses pelatihan model, hingga evaluasi hasil klasifikasi. Setelah kedua model selesai dijalankan, hasil accuracy dari masing-masing algoritma akan dibandingkan untuk mengetahui algoritma mana yang memberikan performa lebih baik pada dataset ini.

## 4.3 Missing Value

Tahap berikutnya adalah menangani data yang memiliki nilai kosong menggunakan node **Missing Value**.

Data yang memiliki nilai kosong dapat menyebabkan proses pelatihan model menjadi kurang optimal. Oleh karena itu, nilai yang kosong perlu diganti agar seluruh data dapat digunakan.

Pada atribut yang bertipe **Integer**, metode yang digunakan adalah **Mean**.

Metode Mean akan mengganti nilai yang kosong menggunakan rata-rata dari seluruh nilai pada kolom tersebut.

Secara matematis, nilai rata-rata dihitung menggunakan rumus berikut.

$$
\bar{x}=\frac{\sum x}{n}
$$

Keterangan:

$$
\bar{x} = \text{rata-rata data}
$$

$$
\sum x = \text{jumlah seluruh data}
$$

$$
n = \text{banyaknya data}
$$

Metode Mean dipilih karena cukup sederhana dan mampu mempertahankan jumlah data tanpa harus menghapus baris yang memiliki nilai kosong.

![](https://cdn.mathpix.com/snip/images/aXH03S4oyP_kb76hDbrGUhgA0w8XsV01vZkuum9zqf0.original.fullsize.png)

## 4.5 Table Partitioner

Setelah proses preprocessing selesai, dataset dibagi menjadi data training dan data testing menggunakan node **Table Partitioner**.

Konfigurasi yang digunakan adalah:

| Parameter | Nilai |
|-----------|--------|
| Data Training | 80% |
| Data Testing | 20% |
| Sampling | Stratified |

Sebanyak **80%** data digunakan sebagai data training untuk melatih model, sedangkan **20%** sisanya digunakan sebagai data testing untuk menguji kemampuan model dalam melakukan prediksi.

Metode **Stratified Sampling** digunakan agar proporsi setiap kelas pada data training dan data testing tetap seimbang. Dengan cara ini distribusi kelas tidak berubah sehingga proses evaluasi model menjadi lebih adil.

![](https://cdn.mathpix.com/snip/images/VkF9rmwZOBWOEscy4x5VKGn3R2GoNSGpWJYmxEZHddg.original.fullsize.png)

## 4.6 Random Forest Learner

Node **Random Forest Learner** digunakan untuk membangun model klasifikasi menggunakan algoritma Random Forest.

Random Forest bekerja dengan membuat banyak pohon keputusan (decision tree). Setiap pohon akan melakukan proses klasifikasi secara independen, kemudian seluruh hasil prediksi digabungkan untuk menentukan kelas akhir.

Pada analisis ini digunakan konfigurasi sebagai berikut.

| Parameter | Nilai |
|-----------|--------|
| Split Criterion | Gini |
| Maximum Tree Depth | 10 |
| Number of Trees | 100 |

Penjelasan parameter:

- **Split Criterion = Gini**

  Digunakan untuk menentukan pemisahan data terbaik pada setiap node pohon keputusan. Semakin kecil nilai Gini, maka pembagian data dianggap semakin baik.

- **Maximum Tree Depth = 10**

  Membatasi kedalaman setiap pohon hingga maksimal 10 level. Pembatasan ini dilakukan agar model tidak terlalu kompleks dan mengurangi kemungkinan overfitting.

- **Number of Trees = 100**

  Model akan membangun sebanyak 100 pohon keputusan. Semakin banyak pohon yang digunakan, biasanya hasil prediksi menjadi lebih stabil karena keputusan diambil berdasarkan mayoritas hasil dari seluruh pohon.

![](https://cdn.mathpix.com/snip/images/a5C0ayGQKWMhlv3ouknukn6ETBfJTSFGMPxbJj04QHc.original.fullsize.png)

## 4.8 Normalizer

Pada algoritma K-Nearest Neighbor, seluruh atribut numerik perlu dinormalisasi terlebih dahulu menggunakan node **Normalizer**.

Metode normalisasi yang digunakan adalah **Min-Max Normalization**.

Rumus Min-Max adalah:

$$
x'=\frac{x-x_{min}}{x_{max}-x_{min}}
$$

Tujuan normalisasi adalah mengubah seluruh nilai atribut ke dalam rentang yang sama sehingga tidak ada atribut yang memiliki pengaruh lebih besar hanya karena memiliki nilai yang lebih tinggi.

Normalisasi sangat penting pada algoritma KNN karena proses klasifikasinya menggunakan perhitungan jarak antar data.

![](https://cdn.mathpix.com/snip/images/0uKXIQSgGoj8IRKwIXr_aT5UNpQ2XBsXAKhqy77ArX0.original.fullsize.png)

## 4.9 Normalizer (Apply)

Node **Normalizer (Apply)** digunakan untuk menerapkan hasil normalisasi dari data training ke data testing.

Hal ini dilakukan agar data testing menggunakan skala yang sama dengan data training sehingga proses prediksi menjadi lebih konsisten.

Tanpa langkah ini, hasil perhitungan jarak pada algoritma KNN dapat menjadi kurang akurat.

![](https://cdn.mathpix.com/snip/images/nHcvbtE9Az6nDz568HwQ7gJa0PYD-DYGgfRXInu3rwU.original.fullsize.png)

## 4.10 K Nearest Neighbor

Node **K Nearest Neighbor** digunakan untuk membangun model klasifikasi menggunakan algoritma KNN.

Pada analisis ini digunakan konfigurasi sebagai berikut.

| Parameter | Nilai |
|-----------|--------|
| Number of Neighbors (K) | 3 |
| Target Column | Grade |

Nilai **K = 3** berarti setiap data baru akan dibandingkan dengan tiga data yang memiliki jarak paling dekat. Selanjutnya kelas yang paling banyak muncul di antara ketiga tetangga tersebut akan menjadi hasil prediksi.

Semakin kecil nilai K, model menjadi lebih sensitif terhadap perubahan data. Sebaliknya, jika nilai K terlalu besar, model dapat kehilangan detail pola yang ada pada data.

## 5. Hasil Analisis Menggunakan Random Forest

Setelah seluruh proses preprocessing selesai, data training digunakan untuk membangun model klasifikasi menggunakan algoritma **Random Forest**. Model yang telah dibuat kemudian digunakan untuk memprediksi data testing sehingga dapat diketahui kemampuan model dalam mengklasifikasikan data berdasarkan atribut yang tersedia.

Pada analisis ini digunakan parameter sebagai berikut.

| Parameter | Nilai |
|-----------|--------|
| Split Criterion | Gini |
| Maximum Tree Depth | 10 |
| Number of Trees | 100 |

![](https://cdn.mathpix.com/snip/images/HHSZNL1en0vT2ZJ6JngKY9TmvQfCJ_q7IjQUQ_sJZRM.original.fullsize.png)

Hasil evaluasi menggunakan node **Scorer** menunjukkan bahwa algoritma **Random Forest** memperoleh nilai **accuracy sebesar 37,9%**.

Random Forest bekerja dengan membangun banyak pohon keputusan, kemudian setiap pohon akan memberikan hasil prediksi. Seluruh hasil tersebut selanjutnya digabungkan menggunakan metode voting untuk menentukan kelas akhir. Cara kerja ini membuat Random Forest lebih stabil dibandingkan hanya menggunakan satu pohon keputusan.

Nilai accuracy sebesar **37,9%** menunjukkan bahwa model masih mengalami kesalahan dalam mengklasifikasikan beberapa data. Hal ini dapat disebabkan oleh karakteristik dataset yang memiliki banyak kelas pada atribut **Grade**, sehingga proses klasifikasi menjadi lebih menantang. Meskipun demikian, Random Forest masih mampu memberikan hasil yang lebih baik dibandingkan algoritma KNN pada dataset ini.

## 6. Hasil Analisis Menggunakan K-Nearest Neighbor (KNN)

Sebelum algoritma KNN dijalankan, seluruh atribut numerik terlebih dahulu dinormalisasi menggunakan metode **Min-Max Normalization**. Proses normalisasi dilakukan agar seluruh atribut berada pada rentang nilai yang sama sehingga perhitungan jarak pada algoritma KNN menjadi lebih akurat.

Parameter yang digunakan pada algoritma KNN adalah sebagai berikut.

| Parameter | Nilai |
|-----------|--------|
| Jumlah Tetangga (K) | 3 |
| Target Column | Grade |

![](https://cdn.mathpix.com/snip/images/XyPer7cSg0peORkjsoWovY4DpWKo_bjjjk_SAEob7fs.original.fullsize.png)

Hasil evaluasi menggunakan node **Scorer** menunjukkan bahwa algoritma **K-Nearest Neighbor (KNN)** menghasilkan nilai **accuracy sebesar 31%**. Artinya, model mampu mengklasifikasikan sekitar **31%** data testing dengan benar.

Pada analisis ini digunakan nilai **K = 3**, sehingga setiap data baru akan dibandingkan dengan tiga data yang memiliki jarak paling dekat. Selanjutnya kelas yang paling banyak muncul dari ketiga tetangga tersebut akan menjadi hasil prediksi.

Hasil tersebut menunjukkan bahwa performa KNN masih berada di bawah Random Forest pada dataset yang digunakan. Hal ini dapat dipengaruhi oleh cara kerja KNN yang mengandalkan perhitungan jarak antar data. Meskipun data telah dinormalisasi menggunakan metode Min-Max, karakteristik dataset yang cukup beragam membuat proses pencarian tetangga terdekat belum mampu menghasilkan prediksi yang optimal.


## 7. Perbandingan Hasil Analisis

Setelah kedua algoritma selesai dijalankan, dilakukan perbandingan berdasarkan nilai **accuracy** yang dihasilkan oleh masing-masing model.

| Algoritma | Accuracy |
|-----------|----------:|
| Random Forest | **37,9%** |
| K-Nearest Neighbor (KNN) | **31%** |

Selisih nilai accuracy dapat dihitung sebagai berikut.

$$
37.9\% - 31\% = 6.9\%
$$

Berdasarkan hasil tersebut, algoritma **Random Forest** memperoleh nilai accuracy yang lebih tinggi dibandingkan algoritma **K-Nearest Neighbor (KNN)** dengan selisih sebesar **6,9%**.

Perbedaan hasil tersebut menunjukkan bahwa Random Forest lebih mampu mengenali pola yang terdapat pada dataset. Hal ini karena Random Forest membangun banyak pohon keputusan dan menggabungkan hasil prediksi dari setiap pohon sehingga keputusan akhir menjadi lebih stabil.

Sementara itu, KNN melakukan klasifikasi berdasarkan jarak antar data. Meskipun proses normalisasi telah dilakukan, performa KNN pada dataset ini masih lebih rendah dibandingkan Random Forest.

Berdasarkan hasil analisis tersebut, algoritma **Random Forest** dapat dikatakan memberikan performa yang lebih baik pada dataset **Higher Education Students Performance Evaluation** karena menghasilkan nilai accuracy yang lebih tinggi dibandingkan algoritma **K-Nearest Neighbor (KNN)**.


## 8. Kesimpulan

Berdasarkan seluruh tahapan analisis yang telah dilakukan menggunakan KNIME, mulai dari proses preprocessing data, pembentukan model, hingga evaluasi hasil klasifikasi, dapat disimpulkan bahwa kedua algoritma dapat digunakan untuk mengklasifikasikan data pada dataset **Higher Education Students Performance Evaluation**.

Hasil analisis menunjukkan bahwa algoritma **Random Forest** memperoleh nilai **accuracy sebesar 37,9%**, sedangkan algoritma **K-Nearest Neighbor (KNN)** memperoleh nilai **accuracy sebesar 31%**.

Berdasarkan perbandingan tersebut, Random Forest memberikan hasil yang lebih baik dengan selisih accuracy sebesar **6,9%**. Hal ini menunjukkan bahwa Random Forest lebih mampu mengenali pola pada dataset dibandingkan KNN yang menggunakan pendekatan berdasarkan jarak antar data.

Selain pemilihan algoritma, proses preprocessing seperti menghapus atribut yang tidak diperlukan, menangani missing value menggunakan metode mean, mengubah tipe data, melakukan normalisasi, serta membagi data menjadi data training dan data testing juga berpengaruh terhadap hasil klasifikasi yang diperoleh.

Secara keseluruhan, dapat disimpulkan bahwa algoritma **Random Forest** lebih sesuai digunakan pada dataset ini dibandingkan **K-Nearest Neighbor (KNN)**. Meskipun nilai accuracy yang diperoleh masih belum terlalu tinggi, Random Forest tetap memberikan hasil yang lebih baik dalam proses klasifikasi. Ke depannya, performa model masih dapat ditingkatkan dengan melakukan seleksi fitur, penyesuaian parameter, atau mencoba algoritma klasifikasi lainnya.

Hasil analisis ini menunjukkan bahwa pemilihan algoritma sangat memengaruhi performa klasifikasi. Oleh karena itu, Random Forest dapat dijadikan salah satu metode yang lebih sesuai untuk digunakan pada dataset Higher Education Students Performance Evaluation.
