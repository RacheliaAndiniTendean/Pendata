# Proyek Missing Values Inputation + Materi Normalisasi data

## 1. Missing Value Imputation

Missing value merupakan kondisi ketika suatu atribut pada dataset tidak memiliki nilai atau terdapat data yang hilang. Kondisi ini dapat terjadi karena berbagai faktor, seperti kesalahan saat proses pengumpulan data, data yang tidak tercatat, maupun kesalahan dalam proses penyimpanan data.

Apabila missing value tidak ditangani dengan baik, hal tersebut dapat mempengaruhi hasil analisis serta menurunkan kualitas model yang digunakan. Oleh karena itu diperlukan metode imputasi untuk mengisi nilai yang hilang sehingga dataset dapat digunakan secara lebih optimal.

Pada analisis ini, proses penanganan missing value dilakukan menggunakan metode **Weighted K-Nearest Neighbor (WKNN)**.


### 2. Metode Weighted K-Nearest Neighbor (WKNN)

Weighted K-Nearest Neighbor (WKNN) merupakan pengembangan dari metode **K-Nearest Neighbor (KNN)** yang memberikan bobot pada setiap data tetangga terdekat.

Pada metode ini, data yang memiliki jarak lebih dekat terhadap data yang memiliki missing value akan memiliki pengaruh yang lebih besar dibandingkan dengan data yang jaraknya lebih jauh.

Tahapan metode WKNN secara umum adalah sebagai berikut:

1. Menghitung jarak antara data yang memiliki missing value dengan data lainnya.
2. Mengurutkan data berdasarkan jarak terkecil.
3. Menghitung bobot setiap tetangga menggunakan rumus bobot.
4. Menghitung nilai imputasi menggunakan rata-rata berbobot.

#### A. Import Library dan Membuat Dataset

Langkah pertama adalah menyiapkan library yang dibutuhkan serta membuat dataset yang akan digunakan dalam proses perhitungan.

#### B. Menghitung Jarak Menggunakan Euclidean Distance

Untuk mengetahui tingkat kemiripan antara dua data digunakan **Euclidean Distance**.

Rumus Euclidean Distance dituliskan sebagai berikut:

$$
d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
$$

Pada kasus ini jarak dihitung menggunakan tiga fitur numerik yaitu:

- temperature
- wind
- precip

Sehingga bentuk rumusnya menjadi:

$$
d = \sqrt{(temperature_i - temperature_t)^2 + (wind_i - wind_t)^2 + (precip_i - precip_t)^2}
$$

Semakin kecil nilai jarak yang diperoleh, maka kedua data tersebut dianggap semakin mirip.

```python
def euclidean(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

**Penjelasan :**

`def euclidean(a, b):` digunakan untuk membuat fungsi bernama euclidean yang menerima dua input berupa data numerik.

`(a - b)` digunakan untuk menghitung selisih nilai antara dua data yang dibandingkan.

`**2` digunakan untuk mengkuadratkan selisih tersebut sehingga semua nilai menjadi positif.

`np.sum((a - b) ** 2)` digunakan untuk menjumlahkan seluruh hasil kuadrat dari selisih fitur.

`np.sqrt(...)` digunakan untuk menghitung akar kuadrat dari jumlah kuadrat tersebut sehingga diperoleh nilai jarak Euclidean.

`return ...` digunakan untuk mengembalikan hasil perhitungan jarak agar dapat digunakan pada proses berikutnya.

#### C. Menentukan Data Target yang Akan Diimputasi

Nilai yang akan diimputasi terlebih dahulu adalah `humidity` pada data ke-8.

Karena Python menggunakan index mulai dari 0, maka data ke-8 berada pada:

$$
index = 7
$$

Untuk menghitung jarak, digunakan tiga fitur numerik sebagai pembanding, yaitu:

$$
X = \{temperature,\ wind,\ precip\}
$$

Ketiga fitur ini dipakai karena nilainya tersedia pada data target dan pada data lain yang akan dijadikan pembanding.

```python
target_index = 7
features = ["temperature", "wind", "precip"]

target = df.loc[target_index, features]
print(target)
```

**Penjelasan :**

`target_index = 7` digunakan untuk menentukan posisi data yang akan diimputasi, yaitu data ke-8.

`features = ["temperature", "wind", "precip"]` digunakan untuk memilih fitur-fitur yang akan dipakai dalam menghitung jarak.

`df.loc[target_index, features]` digunakan untuk mengambil nilai fitur dari data target.

`target = ...` digunakan untuk menyimpan data target agar dapat dibandingkan dengan data lain.

`print(target)` digunakan untuk menampilkan nilai fitur dari data target yang akan diproses.

#### D. Menghitung Jarak Antar Data

Setelah data target ditentukan, langkah berikutnya adalah menghitung jarak antara data target dengan data lain yang memiliki nilai `humidity`.

Rumus yang digunakan yaitu:

$$
d = \sqrt{(temperature_i - temperature_t)^2 + (wind_i - wind_t)^2 + (precip_i - precip_t)^2}
$$

Dalam proses ini:

- data target tidak dibandingkan dengan dirinya sendiri
- data yang nilai `humidity`-nya kosong tidak ikut dihitung
- hasil perhitungan disimpan dalam satu kolom `jarak`

Semakin kecil hasil jaraknya, maka data tersebut semakin dekat dengan data target.

```python
rows = []

for i in df.index:
    if i == target_index:
        continue

    if pd.isna(df.loc[i, "humidity"]):
        continue

    neighbor = df.loc[i, features]
    d = euclidean(target.values, neighbor.values)

    rows.append([
        df.loc[i, "id"],
        df.loc[i, "humidity"],
        d
    ])

jarak_df = pd.DataFrame(rows, columns=["id", "humidity", "jarak"])
print(jarak_df.round(3))
```

**Penjelasan :**

`rows = []` digunakan untuk menyiapkan list kosong sebagai tempat menyimpan hasil perhitungan jarak.

`for i in df.index:` digunakan untuk melakukan perulangan pada seluruh baris data.

`if i == target_index: continue` digunakan untuk melewati data target agar tidak dibandingkan dengan dirinya sendiri.

`if pd.isna(df.loc[i, "humidity"]): continue` digunakan untuk melewati data yang nilai `humidity`-nya kosong.

`neighbor = df.loc[i, features]` digunakan untuk mengambil nilai fitur dari data pembanding.

`euclidean(target.values, neighbor.values)` digunakan untuk menghitung jarak Euclidean antara data target dan data pembanding.

`rows.append([...])` digunakan untuk menyimpan hasil berupa id data, nilai humidity, dan nilai jarak.

`pd.DataFrame(rows, columns=["id", "humidity", "jarak"])` digunakan untuk mengubah hasil perhitungan menjadi tabel.

`print(jarak_df.round(3))` digunakan untuk menampilkan hasil jarak dengan pembulatan 3 angka di belakang koma agar lebih rapi.

#### E. Menghitung Bobot WKNN

Pada metode **Weighted K-Nearest Neighbor**, setiap tetangga diberi bobot berdasarkan nilai jaraknya terhadap data target.

Rumus bobot yang digunakan adalah:

$$
w_i = \frac{1}{d_i}
$$

dengan:

- $w_i$ = bobot data ke-$i$
- $d_i$ = jarak data ke-$i$ terhadap data target

Artinya, data yang jaraknya lebih kecil akan memiliki bobot lebih besar.
Dengan begitu, data yang lebih dekat akan memberi pengaruh lebih besar saat menghitung nilai imputasi.

```python
jarak_df["bobot"] = 1 / jarak_df["jarak"]
print(jarak_df.round(3))
```

**Penjelasan :**

`jarak_df["bobot"] = 1 / jarak_df["jarak"]` digunakan untuk menambahkan kolom baru bernama `bobot`, yang dihitung dari hasil pembagian 1 dengan nilai jarak.

Rumus `1 / jarak` digunakan karena dalam WKNN, data yang lebih dekat harus memiliki pengaruh yang lebih besar.

`jarak_df["bobot"]` digunakan untuk mengakses atau membuat kolom `bobot` pada DataFrame hasil perhitungan.

`print(jarak_df.round(3))` digunakan untuk menampilkan tabel yang sudah dilengkapi dengan nilai bobot.

#### F. Melakukan Perankingan dan Mengambil 3 Tetangga Terbesar

Setelah bobot dihitung, data diurutkan berdasarkan bobot dari yang terbesar ke yang terkecil.

Karena:

$$
w_i = \frac{1}{d_i}
$$

maka hubungan antara jarak dan bobot adalah:

- jarak kecil $\rightarrow$ bobot besar
- jarak besar $\rightarrow$ bobot kecil

Dalam analisis ini digunakan:

$$
k = 3
$$

Artinya, hanya 3 data dengan bobot terbesar yang akan dipakai untuk menghitung nilai imputasi.

```python
ranking_df = jarak_df.sort_values(by="bobot", ascending=False).reset_index(drop=True)
ranking_df["ranking"] = ranking_df.index + 1

k = 3
top3 = ranking_df.head(k)

print("3 ranking terbesar:")
print(top3.round(3))
```

**Penjelasan :**

`jarak_df.sort_values(by="bobot", ascending=False)` digunakan untuk mengurutkan data berdasarkan nilai bobot dari yang terbesar ke yang terkecil.

`ascending=False` berarti pengurutan dilakukan secara menurun.

`.reset_index(drop=True)` digunakan untuk merapikan kembali nomor index setelah data diurutkan.

`ranking_df["ranking"] = ranking_df.index + 1` digunakan untuk membuat kolom `ranking` agar tiap data memiliki peringkat.

`k = 3` digunakan untuk menentukan jumlah tetangga terdekat yang dipakai dalam proses imputasi.

`ranking_df.head(k)` digunakan untuk mengambil 3 data teratas berdasarkan hasil pengurutan.

`top3 = ...` digunakan untuk menyimpan 3 tetangga dengan bobot terbesar.

`print(top3.round(3))` digunakan untuk menampilkan hasil 3 ranking terbesar.

#### G. Menghitung Nilai Imputasi Humidity

Setelah diperoleh 3 tetangga dengan bobot terbesar, langkah berikutnya adalah menghitung nilai imputasi untuk `humidity`.

Rumus imputasi numerik pada metode WKNN adalah:

$$
\hat{x} = \frac{\sum_{i=1}^{k}(w_i \cdot x_i)}{\sum_{i=1}^{k} w_i}
$$

dengan:

- $\hat{x}$ = nilai imputasi
- $w_i$ = bobot tetangga ke-$i$
- $x_i$ = nilai `humidity` pada tetangga ke-$i$
- $k$ = jumlah tetangga terdekat

Prosesnya dilakukan dengan langkah berikut:

1. Mengalikan bobot dengan nilai `humidity`
2. Menjumlahkan seluruh hasil perkalian tersebut
3. Menjumlahkan seluruh bobot
4. Membagi jumlah nilai imputasi dengan jumlah bobot

```python
top3 = top3.copy()
top3["nilai_imputasi"] = top3["bobot"] * top3["humidity"]

jumlah_nilai_imputasi = top3["nilai_imputasi"].sum()
jumlah_bobot = top3["bobot"].sum()

humidity_impute = jumlah_nilai_imputasi / jumlah_bobot

print(top3.round(3))
print(f"Jumlah nilai imputasi = {jumlah_nilai_imputasi:.3f}")
print(f"Jumlah bobot = {jumlah_bobot:.3f}")
print(f"Nilai imputasi humidity data ke-8 = {humidity_impute:.3f}")
```

**Penjelasan :**

`top3 = top3.copy()` digunakan untuk membuat salinan data agar perubahan yang dilakukan tidak memengaruhi DataFrame sebelumnya.

`top3["nilai_imputasi"] = top3["bobot"] * top3["humidity"]` digunakan untuk menghitung kontribusi masing-masing tetangga terhadap nilai imputasi.

`top3["nilai_imputasi"].sum()` digunakan untuk menjumlahkan seluruh nilai imputasi dari 3 tetangga terpilih.

`top3["bobot"].sum()` digunakan untuk menjumlahkan seluruh bobot dari 3 tetangga terpilih.

`humidity_impute = jumlah_nilai_imputasi / jumlah_bobot` digunakan untuk menghitung nilai imputasi akhir sesuai rumus WKNN.

`print(top3.round(3))` digunakan untuk menampilkan tabel top 3 yang sudah memuat kolom `nilai_imputasi`.

`print(f"...{jumlah_nilai_imputasi:.3f}")` digunakan untuk menampilkan jumlah nilai imputasi dengan 3 angka di belakang koma.

`print(f"...{jumlah_bobot:.3f}")` digunakan untuk menampilkan jumlah bobot dengan 3 angka di belakang koma.

`print(f"...{humidity_impute:.3f}")` digunakan untuk menampilkan hasil akhir imputasi humidity.

## 2. Normalisasi Data

Normalisasi data merupakan proses mengubah skala nilai pada atribut numerik agar berada dalam rentang tertentu tanpa mengubah informasi utama dari data tersebut. Normalisasi termasuk tahap **data preprocessing** dalam data mining dan machine learning.

Dalam dataset, nilai antar atribut sering memiliki skala yang berbeda. Misalnya satu atribut bernilai ratusan sedangkan atribut lain bernilai desimal. Perbedaan skala ini dapat mempengaruhi hasil perhitungan, terutama pada algoritma yang berbasis jarak seperti **K-Nearest Neighbor (KNN), Weighted KNN, dan Clustering**.

Dengan melakukan normalisasi, setiap atribut akan berada pada skala yang sama sehingga tidak ada atribut yang mendominasi proses perhitungan.


### A. Macam-Macam Normalisasi Data

Dataset contoh yang digunakan:

| Nilai |
|------|
|227|
|119|
|214|

Nilai minimum = 119  
Nilai maksimum = 227


#### 1. Min-Max Normalization

Min-Max Normalization merupakan metode normalisasi yang mengubah nilai atribut numerik ke dalam rentang tertentu. Rentang yang paling sering digunakan adalah **0 sampai 1**, namun metode ini juga dapat digunakan untuk rentang lain seperti **[-1,1]**.

Rumus umum Min-Max Normalization:

$$
v' = \frac{v - \min(A)}{\max(A) - \min(A)} \times (new_{max} - new_{min}) + new_{min}
$$

Penjelasan rumus:

- $v$ adalah nilai asli data.
- $\min(A)$ adalah nilai minimum dari seluruh data pada atribut tersebut.
- $\max(A)$ adalah nilai maksimum dari seluruh data pada atribut tersebut.
- $new_{min}$ adalah batas minimum dari rentang baru.
- $new_{max}$ adalah batas maksimum dari rentang baru.
- $v'$ adalah nilai hasil normalisasi.

Rumus ini bekerja dengan cara menyesuaikan nilai data dari rentang lama ke rentang baru yang diinginkan.

Jika rentang normalisasi yang digunakan adalah **0 sampai 1**, maka rumusnya dapat disederhanakan menjadi:

$$
v' = \frac{v - \min(A)}{\max(A) - \min(A)}
$$

Artinya nilai minimum akan menjadi **0** dan nilai maksimum akan menjadi **1**.

##### a. Contoh Perhitungan

Untuk nilai **227**

$$
\frac{227 - 119}{227 - 119} = \frac{108}{108} = 1
$$

Untuk nilai **119**

$$
\frac{119 - 119}{227 - 119} = \frac{0}{108} = 0
$$

Untuk nilai **214**

$$
\frac{214 - 119}{227 - 119} = \frac{95}{108} \approx 0.88
$$

Hasil normalisasi:

| Nilai | Hasil |
|------|------|
|227|1.00|
|119|0.00|
|214|0.88|

##### b. Implementasi Python


```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

data = {'nilai':[227,119,214]}

df = pd.DataFrame(data)

scaler = MinMaxScaler()

hasil = scaler.fit_transform(df)

print(hasil)
```

**Penjelasan :**

`from sklearn.preprocessing import MinMaxScaler` digunakan untuk mengimpor class **MinMaxScaler** dari library sklearn yang berfungsi melakukan normalisasi menggunakan metode Min-Max.

`import pandas as pd` digunakan untuk mengimpor library pandas yang digunakan untuk mengelola data dalam bentuk tabel.

`data = {'nilai':[227,119,214]}` digunakan untuk membuat dataset contoh yang berisi nilai numerik.

#### 2. Z-Score Normalization

Z-Score Normalization merupakan metode normalisasi yang mengubah nilai data berdasarkan **rata-rata (mean)** dan **standar deviasi**.

Rumus Z-Score:

$$
v' = \frac{v - \mu}{\sigma}
$$

Penjelasan rumus:

- $v$ adalah nilai asli data
- $\mu$ adalah rata-rata data
- $\sigma$ adalah standar deviasi data
- $v'$ adalah nilai hasil normalisasi

Rumus Mean:

$$
\mu = \frac{1}{n}\sum_{i=1}^{n} v_i
$$

Rumus Standar Deviasi:

$$
\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(v_i - \mu)^2}
$$

##### a. Contoh Perhitungan

Mean:

$$
\mu = \frac{227 + 119 + 214}{3} = 186.67
$$

Standar deviasi:

$$
\sigma \approx 48.27
$$

Hasil Z-Score:

| Nilai | Z-Score |
|------|------|
|227|0.84|
|119|-1.40|
|214|0.57|

##### b. Implementasi Python

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

data = {'nilai':[227,119,214]}

df = pd.DataFrame(data)

scaler = StandardScaler()

hasil = scaler.fit_transform(df)

print(hasil)
```

**Penjelasan :**

`from sklearn.preprocessing import StandardScaler` digunakan untuk mengimpor class **StandardScaler** dari library sklearn yang digunakan untuk melakukan normalisasi menggunakan metode Z-score.

`data = {'nilai':[227,119,214]}` digunakan untuk membuat dataset contoh.

`df = pd.DataFrame(data)` digunakan untuk mengubah dictionary menjadi DataFrame.

`scaler = StandardScaler()` digunakan untuk membuat objek normalisasi menggunakan metode Z-score.

`hasil = scaler.fit_transform(df)` digunakan untuk menghitung nilai mean dan standar deviasi kemudian menormalkan data.

`print(hasil)` digunakan untuk menampilkan hasil normalisasi data.

#### 3. Decimal Scaling Normalization

Decimal Scaling Normalization merupakan metode normalisasi sederhana yang dilakukan dengan menggeser titik desimal nilai data.

Rumus Decimal Scaling:

$$
v' = \frac{v}{10^j}
$$

Penjelasan rumus:

- $v$ adalah nilai asli data
- $j$ adalah jumlah digit dari nilai maksimum data
- $v'$ adalah nilai hasil normalisasi

##### a. Contoh Perhitungan

Nilai maksimum = 227

Karena memiliki **3 digit**, maka

$$
j = 3
$$

Perhitungan:

227 / 1000 = 0.227  
119 / 1000 = 0.119  
214 / 1000 = 0.214  

Hasil normalisasi:

| Nilai | Hasil |
|------|------|
|227|0.227|
|119|0.119|
|214|0.214|

##### b. Implementasi Python

```python
import pandas as pd

data = {'nilai':[227,119,214]}

df = pd.DataFrame(data)

max_val = df['nilai'].max()

j = len(str(max_val))

hasil = df['nilai'] / (10**j)

print(hasil)
```

**Penjelasan :**

`import pandas as pd` digunakan untuk mengimpor library pandas yang digunakan untuk mengelola data dalam bentuk tabel.

`data = {'nilai':[227,119,214]}` digunakan untuk membuat dataset contoh.

`df = pd.DataFrame(data)` digunakan untuk mengubah dictionary menjadi DataFrame.

`max_val = df['nilai'].max()` digunakan untuk mencari nilai maksimum pada dataset.

`j = len(str(max_val))` digunakan untuk menghitung jumlah digit dari nilai maksimum.

`10**j` digunakan sebagai pembagi dalam proses decimal scaling.

`hasil = df['nilai'] / (10**j)` digunakan untuk melakukan normalisasi data.

`print(hasil)` digunakan untuk menampilkan hasil normalisasi.
