#### 4.3.1 Python Script Training Decision Tree

Node **Python Script Training Decision Tree** digunakan untuk melatih model Decision Tree menggunakan data training dari Table Partitioner.

Pada node ini, data training dari KNIME diambil ke Python, kemudian dilakukan preprocessing, pembuatan model, training, dan penyimpanan model menggunakan `pickle`.

Kode lengkap yang digunakan:

```python
import pandas as pd
import os
import pickle
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Ambil data training
df = input_table_1.copy()

# Target
target_col = "Revenue"

# Bersihkan data
df = df.replace("?", np.nan)

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].astype(str).str.strip()

# Kolom numerik
numeric_columns = [
    "Administrative",
    "Administrative_Duration",
    "Informational",
    "Informational_Duration",
    "ProductRelated",
    "ProductRelated_Duration",
    "BounceRates",
    "ExitRates",
    "PageValues",
    "SpecialDay"
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Kolom kategorikal
categorical_columns = [
    "Month",
    "OperatingSystems",
    "Browser",
    "Region",
    "TrafficType",
    "VisitorType",
    "Weekend"
]

for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype(str)

df = df.dropna(subset=[target_col])

# Pisahkan fitur dan target
X_train = df.drop(columns=[target_col])
y_train = df[target_col].astype(str)

# Fitur numerik dan kategorikal
numeric_cols = [col for col in numeric_columns if col in X_train.columns]
categorical_cols = [col for col in categorical_columns if col in X_train.columns]

try:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", encoder)
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

# Model Decision Tree
decision_tree = DecisionTreeClassifier(
    criterion="gini",
    min_samples_leaf=2,
    random_state=42
)

model_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("decision_tree", decision_tree)
])

# Training
model_pipeline.fit(X_train, y_train)

# Simpan model
model_folder = os.path.join(os.path.expanduser("~"), "knime_model")
os.makedirs(model_folder, exist_ok=True)

model_path = os.path.join(model_folder, "decision_tree_revenue_model.pkl")

with open(model_path, "wb") as file:
    pickle.dump(model_pipeline, file)

# Output
output_table_1 = pd.DataFrame({
    "Status": ["Decision Tree berhasil dibuat dan disimpan"],
    "Model Path": [model_path],
    "Jumlah Data Training": [len(df)],
    "Target": [target_col],
    "Jumlah Fitur": [X_train.shape[1]]
})
```

Kode lengkap yang digunakan:

```python
import pandas as pd
import os
import pickle
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Ambil data training
df = input_table_1.copy()

# Target
target_col = "Revenue"

# Bersihkan data
df = df.replace("?", np.nan)

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].astype(str).str.strip()

# Kolom numerik
numeric_columns = [
    "Administrative",
    "Administrative_Duration",
    "Informational",
    "Informational_Duration",
    "ProductRelated",
    "ProductRelated_Duration",
    "BounceRates",
    "ExitRates",
    "PageValues",
    "SpecialDay"
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Kolom kategorikal
categorical_columns = [
    "Month",
    "OperatingSystems",
    "Browser",
    "Region",
    "TrafficType",
    "VisitorType",
    "Weekend"
]

for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype(str)

df = df.dropna(subset=[target_col])

# Pisahkan fitur dan target
X_train = df.drop(columns=[target_col])
y_train = df[target_col].astype(str)

# Fitur numerik dan kategorikal
numeric_cols = [col for col in numeric_columns if col in X_train.columns]
categorical_cols = [col for col in categorical_columns if col in X_train.columns]

try:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", encoder)
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

# Model Random Forest
random_forest = RandomForestClassifier(
    n_estimators=50,
    min_samples_leaf=1,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

model_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("random_forest", random_forest)
])

# Training
model_pipeline.fit(X_train, y_train)

# Simpan model
model_folder = os.path.join(os.path.expanduser("~"), "knime_model")
os.makedirs(model_folder, exist_ok=True)

model_path = os.path.join(model_folder, "random_forest_revenue_model.pkl")

with open(model_path, "wb") as file:
    pickle.dump(model_pipeline, file)

# Output
output_table_1 = pd.DataFrame({
    "Status": ["Random Forest berhasil dibuat dan disimpan"],
    "Model Path": [model_path],
    "Jumlah Data Training": [len(df)],
    "Target": [target_col],
    "Jumlah Fitur": [X_train.shape[1]],
    "Jumlah Tree": [50]
})
```

### 5.1 Confusion Matrix

Confusion matrix digunakan untuk melihat jumlah prediksi benar dan salah pada masing-masing kelas.

Pada dataset ini target `Revenue` memiliki dua kelas:

```text
FALSE = tidak melakukan pembelian
TRUE  = melakukan pembelian
```

Bentuk umum confusion matrix:

```text
Actual FALSE => Predicted FALSE = TN
Actual FALSE => Predicted TRUE  = FP
Actual TRUE  => Predicted FALSE = FN
Actual TRUE  => Predicted TRUE  = TP
```

Keterangan:

```text
TN = True Negative
FP = False Positive
FN = False Negative
TP = True Positive
```

#### a. True Negative

True Negative adalah data dengan nilai asli `FALSE` dan diprediksi `FALSE`.

Artinya model berhasil memprediksi bahwa pengunjung tidak melakukan pembelian.

#### b. False Positive

False Positive adalah data dengan nilai asli `FALSE`, tetapi diprediksi `TRUE`.

Artinya model salah memprediksi bahwa pengunjung melakukan pembelian.

#### c. False Negative

False Negative adalah data dengan nilai asli `TRUE`, tetapi diprediksi `FALSE`.

Artinya model gagal mendeteksi pengunjung yang sebenarnya melakukan pembelian.

#### d. True Positive

True Positive adalah data dengan nilai asli `TRUE` dan diprediksi `TRUE`.

Artinya model berhasil memprediksi bahwa pengunjung melakukan pembelian.

### 5.2 Accuracy

Accuracy digunakan untuk menghitung seberapa banyak prediksi model yang benar dibandingkan seluruh data testing.

Rumus accuracy:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

Semakin tinggi nilai accuracy maka semakin baik performa model secara keseluruhan.

### 5.3 Precision

Precision digunakan untuk melihat ketepatan model saat memprediksi kelas positif.

Rumus precision:

$$
Precision = \frac{TP}{TP + FP}
$$

Precision tinggi menunjukkan bahwa hasil prediksi positif model cukup akurat.

### 5.4 Recall

Recall digunakan untuk melihat kemampuan model dalam menemukan data positif.

Rumus recall:

$$
Recall = \frac{TP}{TP + FN}
$$

Semakin tinggi recall maka model semakin baik dalam mendeteksi data positif.

### 5.5 F1-Score

F1-Score digunakan untuk melihat keseimbangan antara precision dan recall.

Rumus F1-Score:

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

Nilai F1-Score biasanya digunakan ketika ingin melihat performa model secara lebih seimbang.

### 8.1 Analisis Perbandingan

Decision Tree memiliki struktur yang lebih sederhana dan lebih mudah dipahami karena proses klasifikasi hanya menggunakan satu pohon keputusan.

Model ini cocok digunakan ketika interpretasi model menjadi hal penting.

Namun karena hanya menggunakan satu decision tree, model lebih mudah mengalami overfitting terhadap data training.

Pada penelitian ini, Decision Tree menghasilkan accuracy sebesar:

$$
87.95\%
$$

Sedangkan Random Forest menggunakan banyak pohon keputusan sehingga proses prediksi menjadi lebih stabil.

Pada penelitian ini digunakan:

```text
50 decision trees
```

Setiap tree melakukan prediksi terhadap data testing, kemudian hasil akhir ditentukan menggunakan voting.

Random Forest menghasilkan accuracy sebesar:

$$
88.92\%
$$

Hasil tersebut menunjukkan bahwa penggunaan banyak tree mampu meningkatkan performa klasifikasi dibandingkan Decision Tree tunggal.

### 8.2 Alasan Random Forest Lebih Stabil

Random Forest mampu memberikan performa lebih baik karena:

1. Menggunakan banyak decision tree
2. Mengurangi risiko overfitting
3. Menggunakan voting dari seluruh tree
4. Lebih stabil terhadap perubahan data
5. Cocok untuk dataset dengan banyak fitur

Secara konsep, Random Forest tidak hanya mengambil keputusan dari satu pohon, tetapi dari gabungan banyak pohon keputusan.

Konsep voting pada Random Forest dapat dituliskan sebagai:

$$
H(x) = \text{majority vote}(h_1(x), h_2(x), ..., h_n(x))
$$

Keterangan:
- $H(x)$ = hasil akhir prediksi Random Forest
- $h_n(x)$ = hasil prediksi dari tree ke-n
- $n$ = jumlah decision tree
