# Normalisasi Data

## 1. Pengertian Normalisasi Data

Data yang dikumpulkan dalam suatu dataset sering kali belum dapat langsung digunakan oleh algoritma **data mining** atau **machine learning**. Hal ini karena atribut yang tersedia biasanya masih berupa **data mentah (raw attributes)** yang dibuat untuk kebutuhan sistem operasional atau memiliki makna tertentu pada domain asalnya.

Atribut tersebut sering kali belum memiliki karakteristik yang optimal untuk membangun model prediksi yang akurat. Oleh karena itu, diperlukan proses manipulasi atau transformasi data agar atribut memiliki sifat yang lebih sesuai untuk proses pemodelan.

Normalisasi merupakan salah satu teknik transformasi data yang **tidak menghasilkan atribut baru**, tetapi hanya **mengubah skala nilai dari atribut yang sudah ada** sehingga berada pada rentang tertentu.

Normalisasi sangat penting terutama pada algoritma yang sensitif terhadap perbedaan skala data, seperti algoritma berbasis jarak.


## 2. Macam-Macam Normalisasi Data

### 1. Min-Max Normalization

Min-Max Normalization adalah metode normalisasi yang digunakan untuk mengubah nilai atribut numerik ke dalam **rentang tertentu**, biasanya **[0,1]** atau **[-1,1]**.

Metode ini bekerja dengan memetakan nilai asli ke dalam rentang baru berdasarkan nilai minimum dan maksimum dari atribut tersebut.

Rumus Min-Max Normalization:

$$
v' = \frac{v - min(A)}{max(A) - min(A)} \times (new\_max - new\_min) + new\_min
$$

Keterangan:

- \(v\) = nilai asli  
- \(v'\) = nilai setelah normalisasi  
- \(min(A)\) = nilai minimum atribut  
- \(max(A)\) = nilai maksimum atribut  
- \(new\_min\) = batas minimum baru  
- \(new\_max\) = batas maksimum baru  

Metode ini sering digunakan pada algoritma berbasis jarak seperti **K-Nearest Neighbor (KNN)** dan **Clustering**, karena dapat mencegah atribut dengan nilai besar mendominasi perhitungan jarak.

#### Contoh Min-Max Normalization

Data awal:

| Nilai |
|------|
|20|
|30|
|40|

Nilai minimum = 20  
Nilai maksimum = 40  

Hasil normalisasi:

| Nilai | Hasil |
|------|------|
|20|0|
|30|0.5|
|40|1|


### 2. Z-Score Normalization

Z-score normalization atau **standardization** digunakan ketika metode Min-Max kurang efektif, terutama jika terdapat **outlier** atau nilai minimum dan maksimum tidak diketahui.

Metode ini menormalkan data berdasarkan **rata-rata (mean)** dan **standar deviasi**.

Rumus Z-score Normalization:

$$
v' = \frac{v - \mu}{\sigma}
$$

Keterangan:

- \(v\) = nilai asli  
- \(v'\) = nilai setelah normalisasi  
- \(\mu\) = rata-rata (mean) atribut  
- \(\sigma\) = standar deviasi  

Setelah proses normalisasi dilakukan:

- rata-rata data menjadi **0**
- standar deviasi menjadi **1**

#### Rumus Mean

$$
\mu = \frac{1}{n}\sum_{i=1}^{n} v_i
$$

#### Rumus Standar Deviasi

$$
\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(v_i - \mu)^2}
$$

#### Contoh Z-score

Data awal:

| Nilai |
|------|
|20|
|30|
|40|

Mean = 30  

Hasil normalisasi:

| Nilai | Z-score |
|------|------|
|20|-1|
|30|0|
|40|1|


### 3. Decimal Scaling Normalization

Decimal Scaling Normalization adalah metode sederhana yang dilakukan dengan **menggeser titik desimal nilai data** menggunakan pembagian dengan pangkat 10.

Rumus Decimal Scaling:

$$
v' = \frac{v}{10^j}
$$

Keterangan:

- \(v\) = nilai asli  
- \(v'\) = nilai setelah normalisasi  
- \(j\) = bilangan bulat terkecil sehingga nilai maksimum setelah normalisasi < 1 (jumlah digit nilai maksimum pada data)

#### Contoh Decimal Scaling

Data awal:

| Nilai |
|------|
|120|
|450|
|900|

Nilai maksimum = 900  

Karena nilai maksimum memiliki 3 digit, maka \(j = 3\)

Hasil normalisasi:

| Nilai | Hasil |
|------|------|
|120|0.12|
|450|0.45|
|900|0.90|


## 3. Implementasi Normalisasi Menggunakan Python

### 1. Min-Max Normalization Menggunakan Library sklearn

Contoh Min-Max Normalization menggunakan library **sklearn**:


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = pd.DataFrame({
    'umur':[20,30,40],
    'gaji':[4,7,10]
})

scaler = MinMaxScaler()

normalized = scaler.fit_transform(data)

print(normalized)
```

### 2. Z-Score Normalization Menggunakan Library sklearn

Contoh Z-Score Normalization menggunakan library **sklearn**:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data = pd.DataFrame({
    'umur':[20,30,40],
    'gaji':[4,7,10]
})

normalized = scaler.fit_transform(data)
print(normalized)
```

### 3. Decimal Scaling Normalization Menggunakan Function

Contoh Decimal Scaling Normalization menggunakan Function:

```python
import pandas as pd
import numpy as np

data = pd.Series([120,450,900])

def decimal_scaling(x):
    j = len(str(int(x.abs().max())))
    return x / (10**j)

hasil = decimal_scaling(data)

print(hasil)
```
