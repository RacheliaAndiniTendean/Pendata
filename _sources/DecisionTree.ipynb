# Analisa Data Menggunakan Pohon Keputusan (Decision Tree)

## Pendahuluan

Pada proyek ini dilakukan analisis data menggunakan metode Decision Tree atau pohon keputusan. Decision Tree merupakan salah satu metode klasifikasi dalam data mining yang digunakan untuk memprediksi suatu class berdasarkan atribut tertentu.

Metode ini bekerja dengan membentuk struktur seperti pohon yang terdiri dari akar, cabang, dan leaf node. Setiap cabang menunjukkan kondisi tertentu, sedangkan leaf node menunjukkan hasil akhir klasifikasi.

Pada penelitian ini digunakan dataset reservasi hotel dengan target klasifikasi berupa status booking hotel, yaitu:
- Dibatalkan
- Tidak_Dibatalkan

Model klasifikasi dibuat menggunakan node **Decision Tree Learner** pada KNIME dengan metode pengukuran **Gain Ratio**.

## Konsep Dasar Decision Tree

Decision Tree bekerja dengan memilih atribut terbaik untuk membagi data.

Proses pembentukan Decision Tree dilakukan dengan langkah berikut:
1. Sistem membaca data training.
2. Sistem menghitung nilai setiap atribut.
3. Atribut terbaik dipilih sebagai root node.
4. Data dibagi menjadi beberapa cabang.
5. Proses pembagian dilakukan terus sampai data cukup homogen.

Pada penelitian ini digunakan metode **Gain Ratio** untuk memilih atribut terbaik.


## Istilah Penting pada Decision Tree

### 1. Kedalaman Pohon (Tree Depth)

Kedalaman pohon adalah jumlah level dari root node sampai leaf node.

Semakin dalam pohon:
- model semakin kompleks,
- kemungkinan overfitting meningkat.


### 2. Panjang Lintasan (Path Length)

Panjang lintasan adalah jumlah langkah dari root node menuju leaf node.

Contoh lintasan:

```text
tipe_tamu → jenis_pemesanan → tipe_kamar → status_booking
```

Semakin panjang lintasan maka aturan keputusan semakin detail.

### 3. Overfitting

Overfitting terjadi ketika model terlalu mengikuti data training sehingga performa pada data baru menjadi kurang baik.

Ciri overfitting:
- pohon terlalu besar,
- cabang terlalu banyak,
- node terlalu detail.

Untuk mengurangi overfitting digunakan parameter:

```text
Minimum number of records per node = 10
```


### 4. Pruning

Pruning adalah proses penyederhanaan pohon dengan menghapus cabang yang kurang penting.

Tujuan pruning:
- mengurangi overfitting,
- membuat pohon lebih sederhana,
- mempermudah analisis hasil.

Pada penelitian ini digunakan:
```text
Reduced Error Pruning
```

## Entropy

Entropy digunakan untuk mengukur tingkat ketidakpastian atau ketidakteraturan data.

Jika data masih bercampur antara beberapa class, maka nilai entropy tinggi.  
Jika data sudah dominan pada satu class, maka nilai entropy rendah.

Rumus entropy:

$$
Entropy(S) = - \sum_{i=1}^{n} p_i \log_2(p_i)
$$

**Keterangan:**

- $S$ adalah himpunan data.
- $n$ adalah jumlah class.
- $p_i$ adalah proporsi data pada class ke-$i$.
- $\log_2$ berarti logaritma basis 2.

Jika hanya ada dua class, yaitu `Dibatalkan` dan `Tidak_Dibatalkan`, maka rumus entropy dapat ditulis menjadi:

$$
Entropy(S) = -p_{Dibatalkan}\log_2(p_{Dibatalkan}) - p_{Tidak\_Dibatalkan}\log_2(p_{Tidak\_Dibatalkan})
$$

### Penjelasan

Entropy bernilai 0 jika semua data dalam satu kelompok memiliki class yang sama.

Contoh:

```text
Semua data = Tidak_Dibatalkan
```

Maka entropy rendah karena data sudah murni.

Sebaliknya, entropy menjadi tinggi jika data terbagi cukup seimbang antara dua class.

Contoh:

```text
50% Dibatalkan
50% Tidak_Dibatalkan
```

Kondisi tersebut menunjukkan data masih bercampur, sehingga ketidakpastiannya tinggi.



## Information Gain

Information Gain digunakan untuk mengukur seberapa besar penurunan entropy setelah data dibagi berdasarkan suatu atribut.

Rumus Information Gain:

$$
Gain(S,A)=Entropy(S)-\sum_{v \in Values(A)} \frac{|S_v|}{|S|}Entropy(S_v)
$$

**Keterangan:**

- $S$ adalah seluruh data.
- $A$ adalah atribut yang sedang diuji.
- $v$ adalah nilai dari atribut $A$.
- $Values(A)$ adalah semua kemungkinan nilai pada atribut $A$.
- $S_v$ adalah subset data yang memiliki nilai $v$.
- $|S_v|$ adalah jumlah data pada subset $S_v$.
- $|S|$ adalah jumlah seluruh data.
- $Entropy(S_v)$ adalah entropy pada subset data setelah dibagi berdasarkan nilai atribut.

### Penjelasan

Information Gain menghitung seberapa besar atribut dapat mengurangi ketidakpastian data.

Jika suatu atribut mampu membagi data menjadi kelompok yang lebih murni, maka nilai Information Gain akan tinggi.

Contoh:

Jika atribut `tipe_tamu` dapat membagi data menjadi kelompok yang lebih jelas antara `Dibatalkan` dan `Tidak_Dibatalkan`, maka atribut tersebut memiliki nilai Information Gain yang baik.

Namun, Information Gain memiliki kelemahan. Atribut yang memiliki banyak kategori bisa mendapat nilai gain tinggi, meskipun belum tentu benar-benar bagus. Oleh karena itu digunakan Gain Ratio.


## Split Information

Split Information digunakan untuk melihat seberapa besar penyebaran data ketika dibagi berdasarkan suatu atribut.

Rumus Split Information:

$$
SplitInfo(A)= - \sum_{i=1}^{n} \frac{|S_i|}{|S|}\log_2\left(\frac{|S_i|}{|S|}\right)
$$

**Keterangan:**

- $A$ adalah atribut yang diuji.
- $S_i$ adalah subset data ke-$i$ setelah data dibagi berdasarkan atribut.
- $|S_i|$ adalah jumlah data pada subset ke-$i$.
- $|S|$ adalah jumlah seluruh data.
- $n$ adalah jumlah cabang atau jumlah nilai pada atribut tersebut.

### Penjelasan

Split Information digunakan untuk menghitung penyebaran data pada setiap cabang.

Jika suatu atribut membagi data ke dalam banyak cabang kecil, maka nilai Split Information akan memengaruhi hasil Gain Ratio.

Tujuannya adalah agar atribut dengan kategori terlalu banyak tidak langsung dianggap paling baik hanya karena menghasilkan banyak pembagian.


## Gain Ratio

Gain Ratio merupakan pengembangan dari Information Gain.

Gain Ratio digunakan untuk mengurangi bias Information Gain terhadap atribut yang memiliki banyak kategori.

Rumus Gain Ratio:

$$
GainRatio(A)=\frac{Gain(S,A)}{SplitInfo(A)}
$$

**Keterangan:**

- $Gain(S,A)$ adalah nilai Information Gain dari atribut $A$.
- $SplitInfo(A)$ adalah nilai Split Information dari atribut $A$.
- $GainRatio(A)$ adalah nilai akhir yang digunakan untuk menentukan atribut terbaik.

### Penjelasan

Atribut dengan nilai Gain Ratio terbesar akan dipilih sebagai node pada Decision Tree.

Gain Ratio dianggap lebih baik dibandingkan Information Gain karena tidak hanya memperhatikan hasil pembagian data, tetapi juga mempertimbangkan penyebaran data pada setiap cabang yang terbentuk.

Pada penelitian ini, metode Gain Ratio digunakan karena sesuai untuk membangun model Decision Tree pada data kategorikal.

Pada node Decision Tree Learner di KNIME, proses perhitungan Entropy, Information Gain, Split Information, dan Gain Ratio dilakukan secara otomatis oleh sistem untuk menentukan atribut terbaik dalam pembentukan pohon keputusan.

## Confusion Matrix

Confusion Matrix digunakan untuk membandingkan hasil prediksi model dengan data asli.

Tabel confusion matrix:

| Aktual | Dibatalkan | Tidak_Dibatalkan |
|---|---:|---:|
| Dibatalkan | 1819 | 1076 |
| Tidak_Dibatalkan | 1747 | 6241 |

Penjelasan:
- True Positive (TP) = 1819
- False Negative (FN) = 1076
- False Positive (FP) = 1747
- True Negative (TN) = 6241

Nilai diagonal utama menunjukkan prediksi yang benar.

## Perhitungan Accuracy

Accuracy digunakan untuk menghitung tingkat ketepatan model.

Rumus:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN} \times 100\%
$$

Keterangan:
- $TP$ = True Positive
- $TN$ = True Negative
- $FP$ = False Positive
- $FN$ = False Negative

Substitusi nilai:

$$
Accuracy = \frac{1819 + 6241}{1819 + 6241 + 1747 + 1076} \times 100\%
$$

$$
Accuracy = \frac{8060}{10883} \times 100\%
$$

$$
Accuracy \approx 74\%
$$

Accuracy sekitar 74% menunjukkan bahwa model cukup baik dalam melakukan klasifikasi status booking hotel.
